```python
#color by SV (R,G,B)
def get_sv_type(type_str):
    type_str = type_str.strip()
    if "del" in type_str:
        return "255,0,0"
    elif "dup" in type_str:
         # ignore duplications to stay in concordance with ctrl_df['Type]
        return "0,0,255" #blue
    elif "ins" in type_str:
        return "0,0,255"
    elif "inv" in type_str:
        return "0,255,255"
    elif "trans" in type_str:
        return "255,0,255"
    elif "mono" in type_str:
        return "255,0,0"
    elif "tri" in type_str:
        return "0,0,255"
    elif "rear" in type_str:
        return "128,0,128"
    else:
        return(type_str)

#outputting SV colors using python for loops
for type in df.loc[:3,'Type']:
    print(get_sv_type(type))

#adding SV colors column using pandas
#.apply calls pandas to use lambda function
df['Color'] = df.apply(lambda x: get_sv_type(x['Type']), axis=1)

#shows first 5 rows of dataframe
df.head()
```

```python
# popuilate name column
# df["Period"] = df[["Courses", "Duration"]].apply("-".join, axis=1)

# Name
# 	- Size_Type_ISCN_Classification
# 	- Zygosity?
# Prevalence (n patients)


# df['name'] = df[["Size", "Type", "ISCN", "classification"]].apply("_".join, axis=1)
df['name'] = df['Size'].astype(str) + "_" + df["Type"] + "_" + df["ISCN"] + "_" + df["classification"]
```
